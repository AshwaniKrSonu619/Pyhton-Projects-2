```python
df=pd.read_csv("Wine.csv")
df.head()
```

```python
x=df[['Alcohol','Malic_Acid','Ash','Ash_Alcanity','Magnesium','Total_Phenols','Flavanoids','Nonflavanoid_Phenols','Proanthocyanins','Color_Intensity','Hue','OD280','OD280']]
x.head()
```

```python
y=df[['Customer_Segment']]
y
```

# **Scaling Data using Standard Scaler**

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
x_scaled
```

# **Split Data for Training & Testing**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

# **PCA with 95% information**

```python
from sklearn.decomposition import PCA
pca=PCA(0.95)
x_pca=pca.fit_transform(x)
x_pca.shape
```

# **Contribution of these two componnets**

```python
pca.explained_variance_ratio_
print(pca.explained_variance_ratio_)
```

```python

pca=PCA(n_components=2)
x_pca=pca.fit_transform(x)
print(x_pca.shape)
pca.explained_variance_ratio_
```
