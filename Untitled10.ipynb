## Objective:-

**WAP for UNSUPERVISED on mall customer Data Set**

 Theory:-
As we know that K-means is Unsupervised Learning technique in which we got UNLABELLED Data , And in this we try to find pattern of data on the basis Of Similarities in Data points.So In K-Means
Main Objectives are 1-To find Value of K either by ELBOW or Sillhouete 2-to find similarities between Data Points-----using Distance method 3- Repeat the process untill Position of centroids
Changes.
Code:-

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
#%matplotlib inline
```

```python
df=pd.read_csv("Shopping_Mall_Customers.csv")
df.head(10)
```

# **General Observation for deciding value of K in K-means**

```python
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'])
```

```python

k_rng=range(1,10)
sse=[]
for k in k_rng:

 km=KMeans(n_clusters=k)
 km.fit(df[['Annual Income (k$)','Spending Score (1-100)']])
 sse.append(km.inertia_)
sse
```

# Number of Cluster i.e. K by Elbow Method

```python
plt.xlabel('k')
plt.ylabel("Sum of square error")
plt.plot(k_rng,sse)
plt.show()
```

```python
km=KMeans(n_clusters=5)
km
```

```python

y_predicted=km.fit_predict(df[['Annual Income (k$)','Spending Score (1-100)']])
y_predicted
```

```python
df['cluster1']=y_predicted
df.head(50)
```

# implementation of K-Means without Normalizing the data

```python
df1=df[df.cluster1==0]
df2=df[df.cluster1==1]
df3=df[df.cluster1==2]
df4=df[df.cluster1==3]
df5=df[df.cluster1==4]
plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],color='red')
plt.scatter(df2['Annual Income (k$)'],df2['Spending Score (1-100)'],color='blue')
plt.scatter(df3['Annual Income (k$)'],df3['Spending Score (1-100)'],color='green')
plt.scatter(df4['Annual Income (k$)'],df4['Spending Score (1-100)'],color='yellow')
plt.scatter(df5['Annual Income (k$)'],df5['Spending Score (1-100)'],color='orange')
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
```

# **Implementation** **of K-means after Normalization**

**Normalization Of Data**

```python
scaler=MinMaxScaler()
scaler.fit(df[['Annual Income (k$)']])
df[['Annual Income (k$)']]=scaler.transform(df[['Annual Income (k$)']])
scaler.fit(df[['Spending Score (1-100)']])
df[['Spending Score (1-100)']]=scaler.transform(df[['Spending Score (1-100)']])
df
```

```python
km=KMeans(n_clusters=5)
y_predicted=km.fit_predict(df[['Annual Income (k$)','Spending Score (1-100)']])
y_predicted
```

```python
df['cluster']=y_predicted
#df.drop('cluster1',axis='columns',inplace=True)
df
```

# **Implementaion of K-Means**

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
df4=df[df.cluster==3]
df5=df[df.cluster==4]
plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],color='red')
plt.scatter(df2['Annual Income (k$)'],df2['Spending Score (1-100)'],color='blue')
plt.scatter(df3['Annual Income (k$)'],df3['Spending Score (1-100)'],color='green')
plt.scatter(df4['Annual Income (k$)'],df4['Spending Score (1-100)'],color='yellow')
plt.scatter(df5['Annual Income (k$)'],df5['Spending Score (1-100)'],color='black')
```
